```python
for i in range (len(df.itemid)):
    if df.category[i] != 'Handphone & Aksesoris':
         if df.category[i] != 'Elektronik':
                df.drop([i], inplace = True, axis = 0)

df = df.reset_index()
df
```

```python
df.rename(columns = {'discount':'discount(%)'},inplace = True)
df['discount(%)'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
df
```

```python
df['discount(%)'] = df['discount(%)'].fillna(0)
df['brand'] = df['brand'].fillna('null')
df
```

```python
df['name'] = df['name'].astype(str)
df['brand'] = df['brand'].astype(str)
df['category'] = df['category'].astype(str)
df['discount(%)'] = df['discount(%)'].astype(int)
df['curr_price'] = df['curr_price'].astype(str)
```

```python
nulldata = df[df.isna().any(axis = 1)]
nulldata
```
