```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_shopee.csv")
df = df.drop(['index','itemid'], axis = 1)
df
```

```python
df.corr()
```

```python
def is_category(cat):
    if cat == 'Handphone & Aksesoris':
        return '0'
    else:
        return '1'
    
df.category = df.category.apply(is_category)
df
```
